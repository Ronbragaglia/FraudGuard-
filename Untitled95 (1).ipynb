```python
import zipfile
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.metrics import roc_auc_score, precision_recall_curve

zip_file_name = 'archive.zip'
extract_dir = 'extracted_files'

with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

extracted_files = os.listdir(extract_dir)
print("Arquivos extraídos:", extracted_files)

csv_file_name = None
for file in extracted_files:
    if file.endswith('.csv'):
        csv_file_name = file
        break

if csv_file_name is None:
    print("Nenhum arquivo CSV encontrado no arquivo ZIP.")
else:
    csv_path = os.path.join(extract_dir, csv_file_name)

    try:
        data = pd.read_csv(csv_path)
        print("Arquivo CSV carregado com sucesso.")
        print(data.head())

        print(data.isnull().sum())
        print(data.columns)

        data = data.drop(columns=['non_essential_column'], errors='ignore')
        data = data.fillna(0)

        X = data.drop(columns=['Class'])
        y = data['Class']

        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

        def build_autoencoder(input_dim):
            input_layer = Input(shape=(input_dim,))
            encoded = Dense(64, activation='relu')(input_layer)
            encoded = Dense(32, activation='relu')(encoded)
            encoded = Dense(16, activation='relu')(encoded)

            decoded = Dense(32, activation='relu')(encoded)
            decoded = Dense(64, activation='relu')(decoded)
            decoded = Dense(input_dim, activation='sigmoid')(decoded)

            autoencoder = Model(inputs=input_layer, outputs=decoded)
            autoencoder.compile(optimizer='adam', loss='mean_squared_error')
            return autoencoder

        input_dim = X_train.shape[1]
        autoencoder = build_autoencoder(input_dim)

        autoencoder.fit(X_train, X_train, epochs=50, batch_size=32, validation_split=0.2)

        X_test_reconstructed = autoencoder.predict(X_test)
        mse = np.mean(np.power(X_test - X_test_reconstructed, 2), axis=1)

        threshold = np.percentile(mse, 95)

        y_pred = [1 if e > threshold else 0 for e in mse]

        roc_auc = roc_auc_score(y_test, y_pred)
        precision, recall, _ = precision_recall_curve(y_test, mse)

        plt.plot(recall, precision, marker='.')
        plt.title('Precision-Recall Curve')
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.show()

        print(f'ROC AUC: {roc_auc:.2f}')

        plt.hist(mse, bins=50, alpha=0.7, label='MSE')
        plt.axvline(threshold, color='r', linestyle='dashed', linewidth=2, label='Threshold')
        plt.title('Erro de Reconstrução')
        plt.xlabel('Erro')
        plt.ylabel('Número de Amostras')
        plt.legend()
        plt.show()

    except FileNotFoundError:
        print(f"Arquivo {csv_path} não encontrado.")
    except pd.errors.EmptyDataError:
        print(f"O arquivo {csv_path} está vazio.")
    except pd.errors.ParserError:
        print(f"Erro ao parsear o arquivo {csv_path}.")
```
